```python
from control import *
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

# Ex1 : Parametros especificados
- zeta = 0.5, atenuação <= 3 e erro estático de posição nulo

```python
C = 41.58 * TransferFunction([1, 3], [1, 7.47])
C
```

```python
G = TransferFunction([1], [1, 1, 0])
G
```

```python
interval = np.linspace(0, 10, 1000)
t, y = step_response(C*G / (1 + C*G), interval)
```

```python
# Resposta ao degrau
plt.plot(t, y)
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.grid()
```
