# Exemplo 2

```python
ex = TransferFunction([1], [1, 3, 2, 0])
```

```python
ex
```

```python
_ = root_locus(ex, kvect=np.linspace(0, 7, 100), xlim=(-5, 5), ylim=(-5, 5))
```

```python
k = 0.5 # Acima de 0.34 começa a ter comportamento oscilatório
```

```python
x, y = step_response(k*ex/(1 + k*ex))
```

```python
plt.plot(x, y)
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.grid()
```
