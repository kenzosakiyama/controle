```python
import control as co
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

# Integrador

```python
integrator = co.tf([1], [1, 0, 0])
```

```python
integrator
```

```python
interval = np.linspace(0, 10, 1000)
```

```python
t1, y1 = co.impulse_response(integrator, interval)
```

```python
plt.plot(t1, y1)
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.grid()
```
