```python
%matplotlib inline
import control as co
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

- wn interfere na rapidez do sistema,de forma que quanto maior mais rápido e quanto menor mais lento.
- zeta introduz oscilações no sistema quanto menor seu valor (ele modifica o valor da frequência natural). Por outro lado, zeta >= 1 elimina completamente as oscilações.
- Quanto maior o zeta, menor o overshoot do sistema.

# Sub amortecido (0 < zeta < 1)

```python
wn = 5
zeta = 0.3
```

```python
ft = co.tf([wn**2],[1, 2*zeta*wn, wn**2])
```

```python
print(ft)
```

```python
interval = np.linspace(0, 10, 10000)
t, y = co.step_response(ft, interval)
```

```python
plt.plot(t, y)
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.grid()
```

# Sobreamortecido (zeta > 1)

- zetas superiores retardam o sistema.
- O erro é puramente exponencial, e o zeta irá afetar o escalar da exponencial.

```python
wn = 5
zeta = 5
```

```python
ft = co.tf([wn**2],[1, 2*zeta*wn, wn**2])
```

```python
print(ft)
```

```python
interval = np.linspace(0, 10, 10000)
t, y = co.step_response(ft, interval)
```

```python
plt.plot(t, y)
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.grid()
```
