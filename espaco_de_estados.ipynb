```python
%matplotlib inline
import control as co
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

# Especificando o sistema através das matrizes A, B, C e D

- Usando exemplo de circuito elétrico da aula de laboratório 1.
- Sitemas elétricos são rápidos mesmo.


```python
r = 100e-3
R = 20
L = 250e-6
C = 470e-6
```

```python
A = np.matrix(f"{-1/(R*C)}, {-1/C} ; {1/L} , -{r/L}")
B = np.matrix(f"0 ; {1/L}")
C = np.matrix("-1, 0")
D = np.matrix(f"0")
```

```python
RLC = co.StateSpace(A, B, C, D)
```

```python
RLC
```

# Resposta ao degrau

```python
interval = np.linspace(0, 100e-3, 1000)
t, y = co.step_response(RLC, interval)
```

```python
plt.plot(t, y)
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.grid()
```

# Resposta ao impulso

```python
t, y = co.impulse_response(RLC, interval)
```

```python
plt.plot(t, y)
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.grid()
```
