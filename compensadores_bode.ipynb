```python
import control as co
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

# Sem compensação

```python
G = co.TransferFunction([1], [1, 0]) * co.TransferFunction([1], [1, 1]) * co.TransferFunction([1], [1, 2])
```

```python
G
```

```python
_, _, _ = co.bode_plot(G, dB=True, margins=True)
```

# Adicionando compensador proporcional

```python
_, _, _ = co.bode_plot(25 * G, dB=True, margins=True)
```

# Adicionando um compensador avanço-atraso

```python
C = 25 * co.TransferFunction([2, 1], [.24, 1]) * co.TransferFunction([7.14, 1], [59.47, 1])
```

```python
C
```

```python
_, _, _ = co.bode_plot(C * G, dB=True, margins=True)
```
